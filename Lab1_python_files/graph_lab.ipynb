```python
from Graph import *
from graph_structure import *
import matplotlib.pyplot as plt


def deg_plot(graph):
    vertices = graph.get_vertices()
    all_deg = []
    occurrence = {}

    for vertex in vertices:
        all_deg.append(graph.degree(vertex))

    all_deg.sort()
    for deg in all_deg:
        occurrence[deg] = all_deg.count(deg)

    plt.plot(occurrence.keys(), occurrence.values(), color = 'r', linestyle = 'solid',
                marker = 'o',label = "Degree's distribution")
        
    plt.xlabel('Degrees')
    plt.ylabel('Number of nodes')
    plt.title('Distribution of nodes\' degree', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()

def main():
    f = open("C:/Users/diamo/OneDrive - McMaster University/3rd Year (Sep 2022 - April 2023)/First Semester (Fall)/SFWRENG 3XB3 - Binding Theory to Practice/Lab 1/l1-graph-lab/_dataset/london.connections.csv", "r")
    data = (read_input(f))
    dict = (create_weighted_connections(data))
    G = Graph(dict)
    deg_plot(G)

main()
```

# Design Decisions and Analyzing

1. The nodes for the graph which are the stations, are stored in "london.stations.csv" and there it is identified as "id" followed by its properties of longitude and latitude, name, display_name, zone, total_lines and rail. It was decided that it is more convenient to store the "id" as the nodes since each station has a unique "id" and it is easier to refer to as compared to its properties.
Edges were determined to be in "london.connections.csv" as it contains the connection between two stations and provides a weight (time). 
a. It is assumed that since subways usually travel both ways, a train leaving from "station1" to "station2" takes the same time going either way, on the same line. Following this consideration, the graph represented would be an undirected weighted graph (with time being the weight).

2. For the implementation of graph class, it receives a dictionary of extracted information from the .csv files where stations identify as keys and weight as values. This makes them totally independent from the format of the .csv file and if the format is changed, the key and value extraction will still be the same.

3. We made a separate function which gives us our nodes in a form of a dictionary and using that dictionary we decided to make the list of all the edges which are provided by the "london.connections.csv". Additionally, there is a function for the average degree of nodes which essentially takes each node's dictionary and finds the number of connecting stations and takes the average.

- Functions mentioned can be found in Graph.py as num_of_vertices, num_of_edges and avg_degree.

4. For the best path, we decided to give time as the highest priority, then the number of stops and then the distance. The decision was made on the conclusion that most people, when they are trying to reach a destination would prefer the path to take less time as compared to the number of stops and distance. Distance not being our top priority is due to the fact that we are using a subset of London's subway network which is a public transport, so the end user would cares less about the distance they travel as compared to the time they are using private transport methods, such as car, bike, etc. These choices can be seen in Itinerary.py.

- When two different paths have the same time, we check which has lower number of stops between them. If the time and the number of stops are the same, then the path is decided based on the distance each path travels. From our findings, having the exact same distance for two different paths is extremely unlikely and thus that will result in our best path.

# SOLID

To ensure our program follows the SOLID principles, we made sure that each of the major tasks is divided into it's own separate python file/class which gives them single responsibilities. This includes methods for extracting data from the csv files to the algorithms we use to find different paths. Both A* and Dijkstra can work separately without being dependant on one another. Similarly, the design is open for extension and closed for modification. We can see that in the Graph class, where graph_structure can use it's methods by making an object and without breaking the code it can do that for multiple objects of different data sets. No additional methods which are never implemented have been created, the design can handle addtional methods if needed but if they are not required they become useless and hence are removed. Our high level module which is the graph_structure does not depend on modules like Graph and therefore does not result in an error when ran without it. This ensures the program follows dependency inversion principle. The new functions/methods added for the week 2 component follow the same principles.
